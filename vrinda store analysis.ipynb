```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Vrinda Store Data Analysis.xlsx')
```

```python
df.shape
```

```python
df.sample(6)
```

```python
df['Age category'] = df['Age'] .apply(lambda x : 'old' if x >= 50 else 'adult' if x > 25 else 'teen')
```

```python
df.head(3)
```

```python
df.info()
```

```python
df['dayofweek'] = df['Date'].dt.day_name()
```

```python
df['Date']
```

```python
df.sample(5)
```

```python
df['Qty'] = df['Qty'].replace("One",1)
df['Qty'] = df['Qty'].replace("Two",2)
```

```python
df.tail(3)
```

```python
df['month'] = df['Date'].dt.month_name()
```

```python
plt.subplots(figsize = (12,8))
plt.barh(df['month'],df['Amount'] , color = "r")
plt.title("Total amout according to month")
plt.xlabel("Amount")
plt.ylabel("Month")
plt.show()
```

In august there are maximum sales and in december there is minimun sales

```python
plt.subplots(figsize = (12,8))
plt.bar(df['Category'],df['Qty'] , color = "b")
plt.title(" sales of Category by quantity")
plt.xlabel("Category")
plt.ylabel("Qty")
plt.show()
```

```python

ax = plt.subplots(figsize = (12,8))

ax=plt.bar(df['Category'],df['Amount'] , color = "r")
ax=plt.title("sales of Category by amount")
ax=plt.xlabel("Category")
ax=plt.ylabel("Amount")

plt.show()
```

Even kurta has maximum qty sales but set and western dress are more profitable for business on the other hand bottom has minimum sales and quantity both.
